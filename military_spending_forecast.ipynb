LOAD DATA

```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt

# Define the file path (use your dataset file path here)
file_path = '/content/Military data 1988-2023.csv'

# Load the dataset with proper encoding
military_data = pd.read_csv(file_path, encoding='ISO-8859-1')

# Display the first few rows to understand the structure
military_data.head()
```

DATA CLEANING

```python
# Reload the dataset, skipping rows until the actual data starts (adjust as needed)
military_data_cleaned = pd.read_csv(file_path, encoding='ISO-8859-1', skiprows=6)

# Set the first row as the header
military_data_cleaned.columns = military_data_cleaned.iloc[0]
military_data_cleaned = military_data_cleaned[1:].reset_index(drop=True)

# Display the cleaned data
military_data_cleaned.head()
```

```python
# Retain only rows for the United States and China
military_data_cleaned = military_data_cleaned[military_data_cleaned['country'].str.lower().isin(['usa', 'china'])]

# Reset the index for easier handling
military_data_cleaned = military_data_cleaned.reset_index(drop=True)

# Display the refined dataset
military_data_cleaned.head()
```

EXPLORATORY DATA ANALYSIS

```python
# Extract years and spending data for USA and China
years = military_data_cleaned.columns[3:]  # From 1988 onwards
usa_spending = pd.to_numeric(military_data_cleaned.loc[military_data_cleaned['country'] == 'usa', years].iloc[0].values.flatten(), errors='coerce')
china_spending = pd.to_numeric(military_data_cleaned.loc[military_data_cleaned['country'] == 'china', years].iloc[0].values.flatten(), errors='coerce')

# Convert years to numeric for plotting
years_numeric = pd.to_numeric(years, errors='coerce')

# Plot trends
plt.figure(figsize=(12, 6))
plt.plot(years_numeric, usa_spending, label='USA (in USD)', marker='o')
plt.plot(years_numeric, china_spending, label='China (in Yuan)', marker='o')
plt.title('Trends of Military Spending (1988-2023)', fontsize=16)
plt.xlabel('Year', fontsize=14)
plt.ylabel('Spending (Local Currency)', fontsize=14)
plt.legend(fontsize=12)
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```

```python
# Extract and process GDP proportion data
usa_gdp_proportion = pd.to_numeric(military_data_cleaned.loc[military_data_cleaned['country'] == 'usa', years].iloc[0].values.flatten(), errors='coerce')  # Changed iloc[1] to iloc[0]
china_gdp_proportion = pd.to_numeric(military_data_cleaned.loc[military_data_cleaned['country'] == 'china', years].iloc[0].values.flatten(), errors='coerce') # Changed iloc[1] to iloc[0]

# Plot proportion of GDP
plt.figure(figsize=(12, 6))
plt.plot(years_numeric, usa_gdp_proportion, label='USA Military Spending as % of GDP', marker='o')
plt.plot(years_numeric, china_gdp_proportion, label='China Military Spending as % of GDP', marker='o')
plt.title('Proportion of Military Spending in GDP (1988-2023)', fontsize=16)
plt.xlabel('Year', fontsize=14)
plt.ylabel('Proportion of GDP (%)', fontsize=14)
plt.legend(fontsize=12)
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```

```python
# Extract government expenditure proportions (assuming similar structure)
# Use iloc[0] to access the first row (USA data) for government expenditure
usa_gov_expenditure = pd.to_numeric(military_data_cleaned.loc[military_data_cleaned['country'] == 'usa', years].iloc[0].values.flatten(), errors='coerce')

# If you need China's data, use iloc[1]
# china_gov_expenditure = pd.to_numeric(military_data_cleaned.loc[military_data_cleaned['country'] == 'china', years].iloc[1].values.flatten(), errors='coerce')

# Plot proportion of government expenditure
plt.figure(figsize=(12, 6))
plt.plot(years_numeric, usa_gov_expenditure, label='USA Military Spending as % of Govt Expenditure', marker='o')
# You can add China's data to the plot if you extracted it
# plt.plot(years_numeric, china_gov_expenditure, label='China Military Spending as % of Govt Expenditure', marker='o')
plt.title('Proportion of Military Spending in Govt Expenditure (1988-2023)', fontsize=16)
plt.xlabel('Year', fontsize=14)
plt.ylabel('Proportion of Govt Expenditure (%)', fontsize=14)
plt.legend(fontsize=12)
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```

```python
from statsmodels.tsa.arima.model import ARIMA
import numpy as np
# Global spending data
global_spending_values = [
    1730.0, 1705.0, 1633.9, 1340.6, 1281.1, 1251.0, 1177.7, 1144.1, 1155.1,
    1139.8, 1161.3, 1204.1, 1228.7, 1312.1, 1411.3, 1505.8, 1562.9, 1613.2,
    1679.1, 1780.7, 1904.6, 1943.7, 1948.5, 1929.7, 1892.8, 1883.1, 1909.9,
    1914.8, 1934.9, 1989.8, 2073.1, 2148.9, 2163.6, 2241.5, 2393.6, 2443.4
]

# Convert to a NumPy array for modeling
global_spending_array = np.array(global_spending_values)
# Fit ARIMA model
model = ARIMA(global_spending_array, order=(5, 1, 0))  # Order is (p, d, q)
model_fit = model.fit()

# Summary of the ARIMA model
print(model_fit.summary())
# Forecast the next 10 years (2024-2033)
forecast_steps = 10
forecast = model_fit.forecast(steps=forecast_steps)

# Display the forecast values
forecast_years = list(range(2024, 2024 + forecast_steps))
for year, value in zip(forecast_years, forecast):
    print(f"{year}: {value:.2f} USD Billion")
# Combine historical and forecasted values for plotting
years_combined = list(range(1988, 2023 + forecast_steps + 1))
spending_combined = np.concatenate([global_spending_array, forecast])

# Plot the historical and forecasted values
plt.figure(figsize=(12, 6))
plt.plot(years_numeric, global_spending_values, label='Historical Spending', marker='o', color='blue')
plt.plot(forecast_years, forecast, label='Forecasted Spending', marker='o', linestyle='--', color='red')
plt.title('Global Military Spending Forecast (1988-2033)', fontsize=16)
plt.xlabel('Year', fontsize=14)
plt.ylabel('Spending (USD Billion)', fontsize=14)
plt.legend(fontsize=12)
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```

4. Uncertainty Factors
The ARIMA model assumes that historical patterns and relationships will continue. However, global military spending can be influenced by:
Geopolitical Events: Wars, peace treaties, or alliances could disrupt trends.
Economic Conditions: Recession or growth in major economies can affect defense budgets.
Technological Shifts: Investments in advanced or cost-efficient technologies may alter spending patterns.
5. Residual Diagnostics
Non-Normal Residuals:

The Jarque-Bera test indicates non-normality in residuals, suggesting the model may not fully capture all underlying patterns.
This could be addressed by incorporating external factors (e.g., economic or political variables) or exploring alternative models.
Heteroskedasticity:

The variance of residuals is not consistent, which could slightly impact the reliability of longer-term forecasts.
